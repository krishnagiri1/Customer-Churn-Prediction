```python
# 1_EDA.ipynb

#  Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Load Dataset
df = pd.read_csv('D:/projects/Customer-Churn-Prediction/data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

#  Basic Overview
print("Dataset Shape:", df.shape)
df.head()

#  Check for Null Values
print("\nMissing Values:")
print(df.isnull().sum())

#  Data Types & Info
print("\nData Types:")
df.info()

#  Descriptive Statistics
df.describe()

#  Target Variable Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()

#  Correlation Heatmap (only numerical columns)
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=['int64', 'float64'])  # Only numeric columns
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features Only)')
plt.show()

#  Explore Categorical Features vs Churn
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'Contract', 'PaymentMethod']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Churn', data=df)
    plt.title(f'{col} vs Churn')
    plt.xticks(rotation=45)
    plt.show()
```
